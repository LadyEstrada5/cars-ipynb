Marcas con los precios promedio más altos y bajos

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Agrupar por marca y calcular precio promedio

```python
avg_price_by_brand = df.groupby("manufacturer_name")["price_usd"].mean().sort_values()
```

Grafica de marcas con los precios promedio más altos y bajos

```python
plt.figure(figsize=(14, 6))
sns.barplot(x=avg_price_by_brand.index, y=avg_price_by_brand.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title("Precio promedio por marca")
plt.ylabel("Precio promedio (USD)")
plt.xlabel("Marca")
plt.tight_layout()
plt.show()
```

Relación entre cuántos kilómetros tiene un auto (odometer_value) y su precio (price_usd)

```python
plt.figure(figsize=(12, 6))
sns.regplot(
    data=df,
    x="odometer_value",
    y="price_usd",
    scatter_kws={'alpha':0.1, 's':10},
    line_kws={'color': 'red'},
    ci=None
)
plt.title("Tendencia: ¿El kilometraje afecta el precio?")
plt.xlabel("Kilometraje (odómetro)")
plt.ylabel("Precio (USD)")
plt.tight_layout()
plt.show()
```

Crear una nueva columna con tramos de kilometraje


```python
bin_size = 20000  # Agrupar cada 20 mil km
df['km_group'] = (df['odometer_value'] // bin_size) * bin_size
```

Calcular el precio promedio por grupo de kilometraje

```python
grouped = df.groupby('km_group')['price_usd'].mean().reset_index()
```

Grafica de relación entre cuántos kilómetros tiene un auto

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped, x='km_group', y='price_usd', marker='o')
plt.title('Precio promedio según kilometraje agrupado')
plt.xlabel('Kilometraje (agrupado cada 20,000 km)')
plt.ylabel('Precio promedio (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Elegimos algunas marcas populares para mostrar
top_brands = df['manufacturer_name'].value_counts().head(3).index

# Filtramos el DataFrame a esas marcas
filtered = df[df['manufacturer_name'].isin(top_brands)].copy()
filtered['km_group'] = (filtered['odometer_value'] // 20000) * 20000

# Calculamos el promedio por grupo y marca
grouped = filtered.groupby(['manufacturer_name', 'km_group'])['price_usd'].mean().reset_index()

# Graficamos
plt.figure(figsize=(14, 7))
sns.lineplot(data=grouped, x='km_group', y='price_usd', hue='manufacturer_name', marker='o')
plt.title('Precio promedio según kilometraje por marca')
plt.xlabel('Kilometraje (agrupado cada 20,000 km)')
plt.ylabel('Precio promedio (USD)')
plt.grid(True)
plt.legend(title='Marca')
plt.tight_layout()
```
